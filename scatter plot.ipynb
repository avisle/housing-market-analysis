```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("housing.csv")


# Data Visualization
# Scatter plot of median house value vs. median income
plt.scatter(data["median_income"], data["median_house_value"], alpha=0.5)
plt.title("Median House Value vs. Median Income")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
data = pd.read_csv('California_Houses.csv')

# Extract latitude, longitude, and median housing price columns
latitude = data['Latitude']
longitude = data['Longitude']
median_price = data['Median_House_Value']

# Create a scatter plot with median housing price as color intensity
plt.figure(figsize=(12, 8))
scatter = plt.scatter(longitude, latitude, c=median_price, cmap='viridis', marker='o', s=50)

# Add color bar to show the legend
plt.colorbar(scatter, label='Median Housing Price')

# Set labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Median Housing Price by Location')

# Show the plot
plt.show()
```
